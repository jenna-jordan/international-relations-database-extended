```python
# convert date columns to date datatype and drop version column
df['start_date'] = pd.to_datetime(df['start_date'], format = '%Y-%m-%d')
df['start_date2'] = pd.to_datetime(df['start_date2'], format = '%Y-%m-%d')
df['ep_end_date'] = pd.to_datetime(df['ep_end_date'], format = '%Y-%m-%d')
df = df.drop(columns=['version'])
```

## create observation ids

each row/observation is a conflict-year

```python
df['obs_id'] = df['conflict_id'].astype(str) + '-' + df['year'].astype(str)
df
```

## create episode ids

Conflicts have multiple episodes when a conflict calms down to <25 deaths in a year (and therefore is not recorded as an observation) and the flares up again to >25 deaths in a year (and therefore has another observation). This can happen multiple times.

```python
episode_groups = df.groupby('conflict_id').agg({'start_date2': 'nunique'})
```

```python
def make_num_list(x):
    num_list = [1]
    start = 1
    while start < x:
        start += 1
        num_list.append(start)
    return num_list

episode_groups['num_list'] = episode_groups['start_date2'].apply(make_num_list)
episode_groups
```

```python
episode_ids = episode_groups.num_list.apply(pd.Series).reset_index() \
    .melt(id_vars = ['conflict_id']) \
    .drop(columns=['variable']) \
    .dropna() \
    .sort_values(by=['conflict_id', 'value']) \
    .astype({'value': 'int'}) \
    .reset_index(drop=True)

episode_ids['ep_id'] = episode_ids['conflict_id'].astype(str) + "-" + episode_ids['value'].astype(str)
episode_ids = episode_ids.rename(columns={'value':'episode'})
episode_ids = episode_ids.sort_values(by=['conflict_id', 'episode']).reset_index(drop=True)
episode_ids
```

```python
episode_groups2 = df[['conflict_id', 'start_date2']].drop_duplicates().reset_index(drop=True)
episode_groups2['start_date2'] = pd.to_datetime(episode_groups2['start_date2'], format = '%Y-%m-%d')
episode_groups2 = episode_groups2.sort_values(by=['conflict_id', 'start_date2']).reset_index(drop=True)
episode_groups2 = episode_groups2.rename(columns={'conflict_id':'conflict_id2'})
episode_groups2
```

```python
episode_groups3 = pd.concat([episode_ids, episode_groups2], axis=1)
episode_groups3
```

```python
episode_groups3[episode_groups3['conflict_id'] != episode_groups3['conflict_id2']]
```

```python
episode_groups3 = episode_groups3.drop(columns=['conflict_id2'])
```

```python
df = df.merge(episode_groups3, on=['conflict_id', 'start_date2'])
df
```

## Adjust column values and create new columns as needed

shift conflict type to match correlates of war conflict types

```python
df['type_of_conflict'].unique()
```

```python
df['isInternational'] = np.nan
df['isInternational'] [df['type_of_conflict'] == 4] = 1
df['isInternational'] [df['type_of_conflict'] == 3] = 0
df['isInternational'] = df['isInternational'].astype('Int64')
```

```python
type_dict = {1: "Extrasystemic", 2: "Interstate", 3: "Internal", 4: "Internal"}
df['type_of_conflict'] = df['type_of_conflict'].replace(type_dict)

intensity_dict = {1: "Minor", 2: "War"}
df['intensity_level'] = df['intensity_level'].replace(intensity_dict)

cum_intensity_dict = {0: "Minor", 1: "War"}
df['cumulative_intensity'] = df['cumulative_intensity'].replace(cum_intensity_dict)
```

create incompatibility flags

```python
df['incompatibility'].unique()
```

```python
df['incompatibility_isTerritory'] = 1
df['incompatibility_isTerritory'] [df['incompatibility'] == 2] = 0

df['incompatibility_isGovernment'] = 1
df['incompatibility_isGovernment'] [df['incompatibility'] == 1] = 0
```

```python
df
```

create conflict end date based on last episode end date

```python
ongoing = df[(df.year == 2018) & (df.ep_end == 0)].conflict_id.tolist()
```

```python
conflict_enddates = df.groupby('conflict_id').agg({'ep_end_date':'max'})
conflict_enddates = conflict_enddates.reset_index().rename(columns={'ep_end_date':'end_date'})
conflict_enddates['end_date'] [conflict_enddates.conflict_id.isin(ongoing)] = pd.NaT
conflict_enddates
```

```python
df = df.merge(conflict_enddates, on='conflict_id', how='left')
df
```

## isolate polities/actors for PARTICIPANTS table

gwno, side, and side_id do not sync up, so they cannot be merged. Instead, there will be two PARTICIPANTS tables - gwID and ucdpID

```python
df.columns
```

### GW IDs

```python
sideA_gwno_list = df['gwno_a'].str.split(pat=", ", expand = True)
sideA_gwno = pd.concat([df['obs_id'], sideA_gwno_list], axis=1)
sideA_gwno = sideA_gwno.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideA_gwno = sideA_gwno.rename(columns={'value': 'gw_id'}).drop(columns=['variable'])
sideA_gwno['side'] = "A"
sideA_gwno['role'] = "primary"
sideA_gwno
```

```python
sideA2_gwno_list = df['gwno_a_2nd'].str.split(pat=", ", expand = True)
sideA2_gwno = pd.concat([df['obs_id'], sideA2_gwno_list], axis=1)
sideA2_gwno = sideA2_gwno.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideA2_gwno = sideA2_gwno.rename(columns={'value': 'gw_id'}).drop(columns=['variable'])
sideA2_gwno['side'] = "A"
sideA2_gwno['role'] = "secondary"
sideA2_gwno
```

```python
sideB_gwno_list = df['gwno_b'].str.split(pat=", ", expand = True)
sideB_gwno = pd.concat([df['obs_id'], sideB_gwno_list], axis=1)
sideB_gwno = sideB_gwno.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideB_gwno = sideB_gwno.rename(columns={'value': 'gw_id'}).drop(columns=['variable'])
sideB_gwno['side'] = "B"
sideB_gwno['role'] = "primary"
sideB_gwno
```

```python
sideB2_gwno_list = df['gwno_b_2nd'].str.split(pat=", ", expand = True)
sideB2_gwno = pd.concat([df['obs_id'], sideB2_gwno_list], axis=1)
sideB2_gwno = sideB2_gwno.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideB2_gwno = sideB2_gwno.rename(columns={'value': 'gw_id'}).drop(columns=['variable'])
sideB2_gwno['side'] = "B"
sideB2_gwno['role'] = "secondary"
sideB2_gwno
```

```python
gw_participants = [sideA_gwno, sideA2_gwno, sideB_gwno, sideB2_gwno]
participants_gw_df = pd.concat(gw_participants) \
                        .merge(df[['obs_id', 'conflict_id', 'year']], how='left', on='obs_id') \
                        .sort_values(['conflict_id', 'year']).reset_index(drop=True)
participants_gw_df
```

### UCDP IDs

```python
sideA_ucdp_list = df['side_a_id'].str.split(pat=", ", expand = True)
sideA_ucdp = pd.concat([df['obs_id'], sideA_ucdp_list], axis=1)
sideA_ucdp = sideA_ucdp.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideA_ucdp = sideA_ucdp.rename(columns={'value': 'ucdp_id'}).drop(columns=['variable'])
sideA_ucdp['side'] = "A"
sideA_ucdp['role'] = "primary"
sideA_ucdp
```

```python
sideB_ucdp_list = df['side_b_id'].str.split(pat=", ", expand = True)
sideB_ucdp = pd.concat([df['obs_id'], sideB_ucdp_list], axis=1)
sideB_ucdp = sideB_ucdp.melt(id_vars = ['obs_id']).dropna().reset_index(drop=True)
sideB_ucdp = sideB_ucdp.rename(columns={'value': 'ucdp_id'}).drop(columns=['variable'])
sideB_ucdp['side'] = "B"
sideB_ucdp['role'] = "primary"
sideB_ucdp
```

```python
ucdp_participants = [sideA_ucdp, sideB_ucdp]
participants_ucdp_df = pd.concat(ucdp_participants) \
                            .merge(df[['obs_id', 'conflict_id', 'year']], how='left', on='obs_id') \
                            .sort_values(['conflict_id', 'year']).reset_index(drop=True)
participants_ucdp_df
```

## seperate multivalued columns - location, region - that are dependent on CONFLICT_ID

### location

note - clearly the identifiers and the names are not in the same order, so they are not matching up correctly.

```python
df.columns
```

```python
locations = df[['conflict_id', 'gwno_loc']]
locations = locations.drop_duplicates()
locations
```

```python
location_id_list = locations['gwno_loc'].str.split(pat=", ", expand = True)
location_df = pd.concat([locations['conflict_id'], location_id_list], axis=1) \
                .melt(id_vars = ['conflict_id']).dropna() \
                .rename(columns={'value': 'gw_id'}).drop(columns=['variable']) \
                .sort_values(['conflict_id', 'gw_id']).reset_index(drop=True)
location_df
```

### region

```python
df.columns
```

```python
regions = df[['conflict_id', 'region']]
regions = regions.drop_duplicates()
regions
```

```python
region_list = regions['region'].str.split(pat=", ", expand = True)
region_df = pd.concat([regions['conflict_id'], region_list], axis=1) \
                .melt(id_vars = ['conflict_id']).dropna() \
                .rename(columns={'value': 'incompatibility_region'}).drop(columns=['variable']) \
                .sort_values(['conflict_id', 'incompatibility_region']).reset_index(drop=True)

region_dict = {"1": "Europe", "2": "Middle East", "3": "Asia", "4": "Africa", "5": "Americas"}
region_df['incompatibility_region'] = region_df['incompatibility_region'].replace(region_dict)
region_df
```
