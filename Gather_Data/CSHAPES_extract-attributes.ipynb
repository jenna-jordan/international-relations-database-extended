# Extract attribute information from CShapes shapefiles

3 October 2019

Filepath: `Data/CShapes/Raw/cshapes_0.6/`

```python
import shapefile
import pandas as pd
```

```python
sf = shapefile.Reader("../Data/CShapes/Raw/cshapes_0.6/cshapes")
```

```python
sf.shapeType # polygon
```

```python
len(sf)
```

```python
shapes = sf.shapes()
fields = sf.fields
records = sf.records()
```

```python
fields
```

```python
records[1]
```

```python
countries = []
for record in records:
    attributes = {'country_name': record[0],
                  'area': record[1],
                  'capital_name': record[2],
                  'capital_long': record[3],
                  'capital_lat': record[4],
                  'feature_id': record[5],
                  'cow_code': record[6],
                  'cow_startyear': record[7],
                  'cow_startmonth': record[8],
                  'cow_startday': record[9],
                  'cow_endyear': record[10],
                  'cow_endmonth': record[11],
                  'cow_endday': record[12],
                  'gw_code': record[13],
                  'gw_startyear': record[14],
                  'gw_startmonth': record[15],
                  'gw_startday': record[16],
                  'gw_endyear': record[17],
                  'gw_endmonth': record[18],
                  'gw_endday': record[19],
                  'iso_name': record[20],
                  'iso_num': record[21],
                  'iso_alpha2': record[22],
                  'iso_alpha3': record[23]}
    countries.append(attributes)

df = pd.DataFrame(countries)
df
```

```python
df.dtypes
```

```python
df.to_csv('../Data/CShapes/country_shapes.csv', index=False)
```
